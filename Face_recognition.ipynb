```python
import numpy as np
import cv2, os, matplotlib.pyplot as plt
from sklearn.metrics import classification_report

IMG_H, IMG_W = 112, 92
DATASET = "./dataset"

images, labels = [], []
subjects = sorted(os.listdir(DATASET))
subjects = [s for s in subjects if os.path.isdir(f"{DATASET}/{s}")]

for idx, person in enumerate(subjects):
    for fname in sorted(os.listdir(f"{DATASET}/{person}")):
        img = cv2.imread(f"{DATASET}/{person}/{fname}", cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMG_W, IMG_H))
        images.append(img.flatten().astype(np.float64))
        labels.append(idx)

X = np.array(images)
y = np.array(labels)
print(f"Dataset: {X.shape}, {len(subjects)} subjects")
```
